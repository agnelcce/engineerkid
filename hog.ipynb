```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
```

```python

img1 = imread('kl0.PNG')
fd, hog_img1 = hog(img1, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
print(fd)
print(fd.shape)

img2 = imread('kl1.PNG')
fd, hog_img2 = hog(img2, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
print(fd)
print(fd.shape)

img3 = imread('kl2.PNG')
fd, hog_img3 = hog(img3, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
print(fd)
print(fd.shape)

img4 = imread('kl3.PNG')
fd, hog_img4 = hog(img4, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
print(fd)
print(fd.shape)

img5 = imread('kl4.PNG')
fd, hog_img5 = hog(img5, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
print(fd)
print(fd.shape)
print(img1.shape)
print(img2.shape)
print(img3.shape)
print(img4.shape)
print(img5.shape)

plt.subplot(321), plt.imshow(img1)
plt.subplot(322), plt.imshow(img2)
plt.subplot(323), plt.imshow(img3)
plt.subplot(324), plt.imshow(img4)
plt.subplot(325), plt.imshow(img5)
plt.show()
plt.subplot(321), plt.imshow(hog_img1)
plt.subplot(322), plt.imshow(hog_img2)
plt.subplot(323), plt.imshow(hog_img3)
plt.subplot(324), plt.imshow(hog_img4)
plt.subplot(325), plt.imshow(hog_img5)
plt.show()
```
