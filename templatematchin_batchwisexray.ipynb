```python
import cv2
import glob
from skimage.filters import gaussian
from skimage import img_as_ubyte
import numpy as np 
import os 

images_list = []
size = 256
path = "dataset_train/train_set/0Normal/*.*"

for file in glob.glob(path):
    #print(file)
    img=cv2.imread(file)
    img=cv2.resize(img,(size,size))
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img_te=cv2.imread('template.png',0)
    w,h=img_te.shape[::-1]
    res = cv2.matchTemplate(img,img_te,cv2.TM_CCOEFF_NORMED)
    thre=.67
    loc=np.where(res>=thre)
    for i in zip(*loc[::-1]):
        cv2.rectangle(img,i,(i[0]+w,i[1]+h),(255,0,0),2)
    images_list.append(img)
images_list=np.array(images_list)

img_number=0
for image in range(images_list.shape[0]):
                   input_img=images_list[image,:,:]
                   cv2.imwrite("C:/Users/HP/dataset_train/train_template/0Normal_temp/TmpNormal"+str(img_number)+".png",input_img)
                   img_number+=1
```
