```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

img=cv2.imread('gg.jpg',0)

print(img.shape)
img_te=cv2.imread('template.jpg',0)
w,h=img_te.shape[::-1]
print(w)
print(h)
res = cv2.matchTemplate(img,img_te,cv2.TM_CCOEFF_NORMED)
thre=.67
loc=np.where(res>=thre)
for i in zip(*loc[::-1]):
    cv2.rectangle(img,i,(i[0]+w,i[1]+h),(255,0,0),2)
plt.imshow(img)

```
