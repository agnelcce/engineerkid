```python
import cv2
import matplotlib.pyplot as plt 
    
img=cv2.imread("CROP.PNG")
img = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
plt.imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
kernelSizes = [(3, 3), (5, 5), (7, 7) ]
# loop over the kernels sizes
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, kernelSize)
    opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel,iterations = 4)
    plt.imshow(opening)

edged = cv2.Canny(opening,57,135)
plt.imshow(edged)
contours, _ = cv2.findContours(edged,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
#print(contours)
print('Numbers of contours found=', len(contours))
cv2.drawContours(img,contours,-1,(255,0,0),1)
plt.imshow(img)
```
