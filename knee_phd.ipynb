```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('oa1.png',0)
plt.imshow(img)
eq_img=cv2.equalizeHist(img)
plt.hist(eq_img.flat,bins=100,range=(0,255))
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
cl_img=clahe.apply(img)
cv2.imshow("equa",eq_img)
cv2.imshow("clahe",cl_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('oa1.png',0)
clahe = cv2.createCLAHE(clipLimit=7.0,tileGridSize=(8,8))
cl_img=clahe.apply(img)
plt.hist(cl_img.flat,bins=100,range=(0,255))
ret,thresh1=cv2.threshold(cl_img,200,100,cv2.THRESH_BINARY)
ret,thresh2=cv2.threshold(cl_img,200,255,cv2.THRESH_BINARY_INV)
ret,thresh3=cv2.threshold(cl_img,50,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imshow("clahe",cl_img)
cv2.imshow("THRE",thresh1)
cv2.imshow("INV",thresh2)
cv2.imshow("OTS",thresh3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
