```python
import cv2
import numpy as np
```

```python

# Load the knee image
image = cv2.imread('C:/Users/HP/dataset_train/test_template/1/histeq_00.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Smooth the image using Gaussian blur
blurred = cv2.GaussianBlur(gray, (9, 9), 0)

cv2.namedWindow('Original Image', cv2.WINDOW_NORMAL)
cv2.namedWindow('Grayscale Image', cv2.WINDOW_NORMAL)
cv2.namedWindow('Blurred Image', cv2.WINDOW_NORMAL)

cv2.imshow('Original Image',image)
cv2.imshow('Grayscale Image',gray)
cv2.imshow('Blurred Image',blurred)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Define the initial contour points
# For example, [(x1, y1), (x2, y2), ...]
initial_points = np.array([(x1, y1), (x2, y2), ...], dtype=np.int32)
```

```python
# Define the parameters for the Snake algorithm
alpha = 0.1  # Weight of the image energy term
beta = 0.2  # Weight of the gradient energy term
gamma = 0.5  # Weight of the continuity energy term
max_iterations = 200  # Maximum iterations for convergence

# Apply the Snake algorithm
snake = cv2.createSnakeImage(gray, initial_points, alpha, beta, gamma, coeff_usage=cv2.CV_CHAIN_APPROX_SIMPLE)

# Perform iterations to refine the contour
for _ in range(max_iterations):
    cv2.cvSnakeImage(snake, blurred, alpha, beta, gamma, coeff_usage=cv2.CV_CHAIN_APPROX_SIMPLE)

# Convert the snake to a polygonal approximation
contour = snake[:, 0, :].astype(int)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.segmentation import active_contour

# Load the image
image = plt.imread('C:/Users/HP/dataset_train/test_template/1/histeq_00.png')

# Convert the image to grayscale
gray = rgb2gray(image)

# Define the initial contour points
# For example, [(x1, y1), (x2, y2), ...]
initial_points = np.array([(54,157),
(67, 171),
(87, 173),
(111, 163),
(134, 147),
(155, 160),
(176, 168),
(198, 162),
(207, 179),
(193, 181),
(149, 168),
(137, 169),
(123, 160),
(103, 180),
(69, 185),
(58, 185)], dtype=np.int32)

# Apply the active contour algorithm
snake = active_contour(gray, initial_points, alpha=0.1, beta=0.2, gamma=0.5,coordinates='rc')

# Display the results
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image)
ax.plot(snake[:, 0], snake[:, 1], '-r', lw=2)
plt.show()
```
