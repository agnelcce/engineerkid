```python
import cv2
import glob
from skimage.filters import gaussian
from skimage import img_as_ubyte
import numpy as np 
import os 

images_list = []
size = 256
path = "dataset_train/train_set/0Normal/*.*"

for file in glob.glob(path):
    print(file)
    img=cv2.imread(file)
    img=cv2.resize(img,(size,size))
    images_list.append(img)

images_list=np.array(images_list)

img_number=0
for image in range(images_list.shape[0]):
                   input_img=images_list[image,:,:]
                   in_img=cv2.cvtColor(input_img,cv2.COLOR_BGR2LAB)
                   l,a,b=cv2.split(in_img)
                   equ0=cv2.equalizeHist(l)
                   #smoothened_image=img_as_ubyte(gaussian(input_img,sigma=2,mode='constant',cval=0.0))
                   cv2.imwrite("dataset_test/histeq_0/Doubtful"+str(img_number)+".png",equ0)
                   img_number+=1
```
