```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('kl0.png',0)
dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum =20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))
plt.subplot(141),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('kl0.png',0)
edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2 as cv
img = cv.imread('kl0.png',0)
# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
cv.imwrite('clahe_2.jpg',cl1)
f=cv.imread('clahe_2.jpg',0)
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.subplot(1,2,2),plt.imshow(f,cmap = 'gray')
```

```python
img1 = cv2.resize(f,(1500,1000))
img1_gray = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
ret,t_img1 = cv2.threshold(img1_gray,170,255,cv2.THRESH_BINARY)
kernel = np.ones((5,5),np.uint8)
t_img1 = cv2.morphologyEx(t_img1, cv2.MORPH_OPEN, kernel)

edges = cv2.Canny(t_img1,100,200)
plt.subplot(121),plt.imshow(f,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
